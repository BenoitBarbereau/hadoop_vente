##  Obtenir le profit total obtenu par région du monde.

```python
def region(reader):
    region = []
    for row in reader:
        region.append(row[0])
    return region
        
```

```python
def country(reader):
    country = []
    for row in reader:
        country.append(row[1])
    return country
```

```python
def item_t(reader):
    item_t = []
    for row in reader:
        item_t.append(row[2])
    return item_t
```

```python
def sales_c(reader):
    sales_c = []
    for row in reader:
        sales_c.append(row[3])
    return sales_c
```
